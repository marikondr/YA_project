### Количество сделанных звонков и израсходованных минут разговора по месяцам

```python
df_calls['month'] = pd.DatetimeIndex(df_calls['call_date']).month # Добавление столбца с номером месяца
```

```python
# Создание сводной таблицы с кол-вом звоноков и продолжительностью.
df_calls['duration'] = np.ceil(df_calls['duration']) # Округление до минут
df_users_calls = pd.pivot_table(df_calls, index=['user_id', 'month'], values='duration', aggfunc=['count','sum'])
df_users_calls = df_users_calls.rename({'count': 'number_of_calls', 'sum': 'minutes_sum'}, axis=1) 
#df_users_calls.columns = ['number_of_calls', 'minutes_sum']
#df_users_calls['minutes_sum'] = df_users_calls['minutes_sum'].apply(math.ceil) # Округление до минут
df_users_calls['minutes_sum'] = df_users_calls['minutes_sum'].astype('int') 
df_users_calls
```

### Количество отправленных сообщений по месяцам

```python
df_messages['month'] = pd.DatetimeIndex(df_messages['message_date']).month # Добавление столбца с номером месяца
```

```python
df_users_messages = pd.pivot_table(df_messages, index=['user_id', 'month'], values='id', aggfunc='count')
df_users_messages.columns = ['number_of_messages']
df_users_messages['number_of_messages'] = df_users_messages['number_of_messages'].astype('int')
df_users_messages
```

### Объем израсходованного интернет-трафика по месяцам


```python
df_internet['month'] = pd.DatetimeIndex(df_internet['session_date']).month # Добавление столбца с номером месяца
```

```python
df_users_internet = pd.pivot_table(df_internet, index=['user_id', 'month'], values='mb_used', aggfunc='sum')
df_users_internet.columns = ['traffic_sum']
df_users_internet['traffic_sum'] = np.ceil(df_users_internet['traffic_sum'])
#df_users_internet['traffic_sum'] = df_users_internet['traffic_sum'].apply(math.ceil)
df_users_internet['traffic_sum'] = df_users_internet['traffic_sum'].astype('int')
df_users_internet
```

```python
data = pd.concat([df_users_calls, df_users_messages, df_users_internet], ignore_index=True, axis=1) # Объединение таблиц
data.columns = ['number_of_count', 'calls_sum', 'number_of_messages', 'traffic_sum'] # Переименование столбцов
data['id'] = data.index.get_level_values('user_id')  # Добавление столбца id
data['month'] = data.index.get_level_values('month') # Добавление столбца month

# Как выяснилось ранее не все пользователи используют звонки или смс или интернет.
# Поэтому заменяю NaN заменю на 0 - это не повлияет на итог.

data['traffic_sum'] = data['traffic_sum'].fillna(0)
data['number_of_messages'] = data['number_of_messages'].fillna(0)
data['calls_sum'] = data['calls_sum'].fillna(0)
data['number_of_count'] = data['number_of_count'].fillna(0)
```

```python
data = data.merge(df_users[['user_id','tariff']], on='user_id') # Добавлю столбец с тарифом
data = data.rename({'tariff': 'tariff_1'}, axis=1) 
```

```python
# Создание переменных по каждому тарифу
# Тариф Smart переменные
smart_rub_in_month = df_tariffs.loc[0, 'rub_monthly_fee'] # Стоимость тарифа в месяц
smart_min  = df_tariffs.loc[0, 'minutes_included'] # Кол-во включенных минут в месяц
smart_messages = df_tariffs.loc[0, 'messages_included'] # Кол-во включенных сообщений в месяц
smart_internet = df_tariffs.loc[0, 'mb_per_month_included'] / 1024 # Кол-во включенных ГБ в месяц
smart_extra_min = df_tariffs.loc[0, 'rub_per_minute'] # Цена в рублях за минуту сверх тарифа
smart_extra_mes = df_tariffs.loc[0, 'rub_per_message'] # Цена в рублях за сообщение сверх тарифа
smart_extra_int = df_tariffs.loc[0, 'rub_per_gb'] # Цена в рублях за ГБ сверх тарифа

# Тариф Ultra переменные
ultra_rub_in_month = df_tariffs.loc[1, 'rub_monthly_fee'] # Стоимость тарифа в месяц
ultra_min  = df_tariffs.loc[1, 'minutes_included'] # Кол-во включенных минут в месяц
ultra_messages = df_tariffs.loc[1, 'messages_included'] # Кол-во включенных сообщений в месяц
ultra_internet = df_tariffs.loc[1, 'mb_per_month_included'] / 1024 # Кол-во включенных ГБ в месяц
ultra_extra_min = df_tariffs.loc[1, 'rub_per_minute'] # Цена в рублях за минуту сверх тарифа
ultra_extra_mes = df_tariffs.loc[1, 'rub_per_message'] # Цена в рублях за сообщение сверх тарифа
ultra_extra_int = df_tariffs.loc[1, 'rub_per_gb'] # Цена в рублях за ГБ сверх тарифа

# Создание функции для расчета ежемесячной выручки с каждого пользователя по двум тарифам 

def tariff_all(row):
    if row['id'] in df_tariff_smart_keys:                            
        if row['calls_sum'] <= smart_min:                                            # минуты не превышены
            minutes = 0  
        if row['calls_sum'] > smart_min:                                             # превышение минут
            minutes = ((row['calls_sum'] - smart_min)*smart_extra_min)
        if row['traffic_sum'] <= smart_internet:                                     # интернет не превышен
            internet = 0
        if row['traffic_sum'] > smart_internet:                                      # интернет  превышен
            internet = ((row['traffic_sum'] - smart_internet)*smart_extra_int)
        if row['number_of_messages'] <= smart_messages:                              # смс не превышены 
            messages = 0
        if row['number_of_messages'] > smart_messages:                               # смс не превышены
            messages = ((row['number_of_messages'] - smart_messages)*smart_extra_mes)          
        return minutes + messages + internet+smart_rub_in_month
    elif row['id'] in df_tariff_ultra_keys:                                       
        if row['calls_sum'] <= ultra_min:
            minutes = 0
        if row['calls_sum'] > ultra_min:
            minutes = ((row['calls_sum'] - ultra_min)*ultra_extra_min )
        if row['number_of_messages'] <= ultra_messages:
            messages = 0
        if row['number_of_messages'] > ultra_messages:
            messages = ((row['number_of_messages'] - ultra_messages)*ultra_extra_mes)
        if row['traffic_sum'] <= ultra_internet:
            internet = 0
        if row['traffic_sum'] > ultra_internet:
            internet = ((row['traffic_sum'] - ultra_internet)*ultra_extra_int )
        return minutes + messages + internet+ultra_rub_in_month
    
data['total'] = data.apply(tariff_all,axis=1) # Создание столбца с общей суммой 
display(data.head(20))
```
