```python
def uniq_columns(data_frame): # Создание общей функции для вывода уникальных значений по каждому столбцу 
    for column in data_frame.columns:
        print('Уникальные значения столбца', column)
        print(data_frame[column].unique())
        print('Количество NAN столбца', column)
        print(data_frame[column].isna().sum())
```

**Одинаковые строчки с пропусками в `name` и `genre`**

Считаю, что можно удалить эти строки, т.к. их всего 2 и данные в них не подлежат восстановлению.

**Пропуски в столбце `year_of_release`**

```python
data_new['year_of_release'].isna().sum()
```

```python
'{:.1%}'.format((data_new[data_new['year_of_release'].isna()].shape[0]/data_new['year_of_release'].shape[0]))
```

1.6 % данных в которых не указан год, эти игры для анализа нам не нужны и учитывая их малый процент, считаю можно их удалить.

```python
data_new = data_new[data_new['year_of_release'].notna()]
data_new.info()
```

**Пропуски в столбце `user_score`**

Пропуски в столбце `user_score` и `critic_score ` заменить невозможно, т.к. все оценки индивидуальны. Аббревиатуру "tbd" заменю на NaN, т.к. она означает, недостаточное количество оценок для выставления общей.

```python
data_new['user_score']=data_new['user_score'].replace({'tbd':np.NaN})
data_new['user_score'].sort_values().unique()
```

# Проверка гипотез

## Гипотеза 1. Средний рейтинг пользователей платформ Xbox One и PC одинаковые.

Проверка статистических данных пользователей Xbox One и PC.

```python
user_xone = current_period.query('platform == "XOne"')
user_xone = user_xone[user_xone['user_score'].notna()]
user_xone['user_score'].describe()
```

```python
user_pc = current_period.query('platform == "PC"')
user_pc = user_pc[user_pc['user_score'].notna()]
user_pc['user_score'].describe()
```

Средние значения довольно близки.
<br>

Проверка гипотезы. Будет использован тест Стюдента.
<br>
Нулевая гипотеза - средние пользовательские рейтинги платформ Xbox One и PC равны.
<br>
Альтернативная гипотеза - средние пользовательские рейтинги платформ Xbox One и PC не равны.
<br>
Коэффициент alpha будет равен 0.01.

```python
alpha = 0.01

results = st.ttest_ind(user_xone['user_score'], user_pc['user_score'])

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

## Гипотеза 2. Средние рейтинг пользователей жанров Action и Sports разные

Проверка статистических данных по жанрам Action и Sports.

```python
user_action = current_period.query('genre == "Action"')
user_action = user_action[user_action['user_score'].notna()]
user_action['user_score'].describe()
```

```python
user_sports = current_period.query('genre == "Sports"')
user_sports = user_sports[user_sports['user_score'].notna()]
user_sports['user_score'].describe()
```

Средние значения отличаются.
<br>

Проверка гипотезы. Будет использован тест Стюдента.
<br>
Нулевая гипотеза - средние пользовательские рейтинги жанров Action и Sports равны.
<br>
Альтернативная гипотеза - средние пользовательские рейтинги жанров Action и Sports не равны.
<br>
Коэффициент alpha будет равен 0.01.

```python
alpha = 0.01

results = st.ttest_ind(user_action['user_score'], user_sports['user_score'])

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**
<br>

Первая гипотеза о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые - подтвердилась. Это означает, что пользователи ставят примерно одинаковые оценки играм на плаформах PC и Xbox One. 
<br>

Вторая гипотеза о том, что средние пользовательские рейтинги жанров Action и Sports разные - не подтвердилась. Средние этих выборок различаются. Это говорит о том, что пользователи ставят разные оценки играм с жанрами Action и Sports.
